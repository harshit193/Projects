```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('./BostonHousing.csv')
df.info()
df.head()
```

```python
df.isnull().sum()
```

```python
df.fillna({'rm':df['rm'].mean()},inplace=True)
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.describe()
```

```python
x = df.iloc[:,:-1]
y = df['medv']
```

```python
x.describe()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
x = pd.DataFrame(ss.fit_transform(x),columns=x.columns)
x.head()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

## using averaging and weighted averaging ensemble ml technique

```python
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.ensemble import VotingRegressor
```

```python
li = [('lr',LinearRegression()),('knn',KNeighborsRegressor()),('dtr',DecisionTreeRegressor(max_depth=3)),('svr',SVR(kernel='poly',degree=6))]
vr = VotingRegressor(estimators=li, weights=[4,5,17,6])
vr.fit(x_train, y_train)
```

```python
vr.score(x_train, y_train), vr.score(x_test, y_test)
```

## Using random forest 

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf = RandomForestRegressor(n_estimators=100, max_depth=4)
rf.fit(x_train, y_train)
rf.score(x_train, y_train), rf.score(x_test, y_test)
```
